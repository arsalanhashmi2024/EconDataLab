```python
def is_unique(s):
    a = s.to_numpy() # s.values (pandas<0.24)
    return (a[0] == a).all()
```

```python
# for i in range()
#     
name = name
df_dict = {}


            
        


for i in range(len(tables)):
    del df
    df = tables[i].df.copy()
    
    for col in df.columns:
        df[col] = df[col].replace('-', '')
#         df[col] = df[col].replace('', )
        
    df.columns=[str(i) for i in range(len(df.columns))]

    
    

    blank_cols = [col for col in df.columns if ((df[col].map(lambda x: x.strip()) == '').all())]
    df = df.drop(blank_cols,
            axis=1,
        )
    
    display(df.head(20))
    print('Trim Top')
    b = input()
    print('Trim Max')
    a = input()
    a = int(a) + 1
    b = int(b) + 1
   
    cols_df = df.iloc[b:a-1]
    display(cols_df)
    colss = [cols_df[c].str.cat(sep=' ') if c!='' else x for x,c in zip(range(len(cols_df.columns)),cols_df.columns)]
    
    df.columns = colss
    df = df.iloc[a:]



    
        

        
#     display(df)

    
    
    df_dict[i] = df
    
    print("this is the cleaned")
    empty_cols = [col for col in df.columns if is_unique(df[col])==True]
    # Drop these columns from the dataframe
    df = df.drop(empty_cols,
            axis=1,
        )
    
    
    display(df.head(10))
#     print(name)
    df.to_csv(str(name)+str(i)+'_camelot.csv',index=False)
    time.sleep(10)
    
    clear_output(wait = True)
    
```
